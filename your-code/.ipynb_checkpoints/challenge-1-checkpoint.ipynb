```python


main_repo = 'ironhack-datalabs/datamad1020-rev'
url = f'https://api.github.com/repos/{main_repo}/forks'


```

```python
response = requests.get(url)

if response.status_code == 200:
    forks = response.json()
    languages_set = set()

    for fork in forks:
        
        languages_url = fork['languages_url']
        
        
        languages_response = requests.get(languages_url)

        if languages_response.status_code == 200:
            languages_data = languages_response.json()
            languages_set.update(languages_data.keys())

    languages_list = list(languages_set)
    print(languages_list)
else:
    print(f"Failed to retrieve forks. Status Code: {response.status_code}")
```
